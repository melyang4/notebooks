## Take log file of D and put D together with lat/long

```python

##GET LATLONGS
lpd="/mnt/solexa/mel_yang/bronzeage/Dstats/"
fh="ind_A_v13_TBH5"
ancind="Boshan"
p4="Mbuti"
dir1="D_ANC_EAS_EAS_P4/"
latlongs={}
latlongfile=open("%s%s%s.pd.D_%s_Tianyuan_EAS_%s.map" % (lpd,dir1,fh,ancind,p4),'r')
for line in latlongfile: 
    x=line.split()
    latlongs[x[0]]=(float(x[1]),float(x[2]))
#latlongs["LiangDao"]=(26,121) ##Put just above Atayal
#latlongs["Xinyi"]=(23,105.3) ##Put just below Longlin
Dpd="/mnt/solexa/mel_yang/neolithic/Dstats/"   
subset="ALLEAS_NEAS_Longlin_Mbuti"
north="Boshan"

neas=['HDYM1','Boisman_MN','Bbdong','Boshan','Linzi','Daur','Hezhen','Mongola','Ulchi','Xibo']
for north in neas:
    newmapfile=open("%s%s.D.all.%s.north_%s.map" % (Dpd,fh,subset,north),'w')
    Dfile=open("%s%s.D.all.%s.log" % (Dpd,fh,subset),'r')
    for line in Dfile:
        if "result" not in line: continue
        x=line.split()
        if x[2] != north: continue
        if x[1] not in latlongs.keys(): continue
        D=float(x[5])
        Z=float(x[7])
        if Z<0: continue ##Place Z cutoff of 2
        ancpd="anc" if x[1] in ["LiangDao","Xinyi"] else "pd"
        newmapfile.write("%s\t%s\t%s\t%.4f\t%s\t%.4f\n" % (x[1], latlongs[x[1]][0], latlongs[x[1]][1], D, ancpd, Z))
    newmapfile.close()

    newreffile=open("%s%s.D.all.%s.north_%s.ref" % (Dpd,fh,subset,north),'w')
    newreffile.write("%s\t%s\t%s\t%.4f\n" % ("Longlin", 24.8, 105.3, -5))
    newreffile.close()
```

## Plotting D against lat or long

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math

Dpd="/mnt/solexa/mel_yang/neolithic/Dstats/"
fh="ind_A_v13_TBH5"
north="Daur"
subset="ALLEAS_NEAS_Longlin_Mbuti"
outname="%s%s.D.all.%s.north_%s" % (Dpd,fh,subset,north)
myfile=open(outname+".map","r")
pops,lats,longs,Ds,Zs=[],[],[],[],[]
for line in myfile:
    x=line.split()
    if x[0]=="Xinyi": continue
    pops.append(x[0])
    lats.append(float(x[1]))
    longs.append(float(x[2]))
    Ds.append(float(x[3]))
    Zs.append(float(x[5]))
mytuple=zip(longs,lats,pops,Ds,Zs)
cutoff=sorted(Ds)[::-1][:40][-1]
cutoffinds=[i[2] for i in mytuple if i[3] < cutoff]

maxDind=Ds.index(max(Ds))
centerlong=mytuple[maxDind][0]
centerlat=mytuple[maxDind][1]

inds = [p for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]
ys1 = [d for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]
newlongs = [o for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]
newlats = [a for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]
se = [1.96*(d/z) for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]    
dists=[math.sqrt(abs(o-centerlong)**2 +abs(a-centerlat)**2) for (o,a,p,d,z) in sorted(mytuple) if p not in cutoffinds]    
dists=[i/max(dists) for i in dists]
xs1=dists

LDind=inds.index("LiangDao")

fig = plt.figure(figsize=(17,4))
titles=["D vs Distance from LD, Normalized", "D vs Longitude", "D vs Latitude"]
#for ind1,i in enumerate(xs1): print xs1[ind1],inds[ind1], ys1[ind1]

for plotind,xs1 in enumerate([dists,newlongs,newlats]):
    ax1=fig.add_subplot(1,3,plotind+1)
    #ax1.plot(xs1,ys1,'ro',alpha=0.5)
    ax1.errorbar(xs1, ys1, yerr=se, fmt='ro')
    #ax1.plot(xs1[,ys1,'bo',alpha=0.5)
    ax1.errorbar(xs1[LDind], ys1[LDind], yerr=se[LDind], fmt='bo')
    ax1.set_ylim(-0.01,0.04)
    #ax1.set_xlim(-2500,48000)
    #ax1.set_xlim(48000,-2500)
    ax1.set_xlabel('Longitude')
    ax1.set_ylabel("D")
    #ax1.text(50000,-0.03,labels[ind],fontsize=17)
    plt.title(titles[plotind])
    ax1.axhline(y=0, ls='solid', c='grey',alpha=0.3)
    #ax1.axhline(y=cutoff[ind], ls='dashed', c='grey',alpha=0.3)
    #ax1.axhline(y=maxcutoff[ind], ls='dashed', c='grey',alpha=0.3)
    print
```
