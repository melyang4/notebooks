```python
import projmodules as pm

seas=['Japanese','Korean','Han_NChina','Han_Shandong','Han_HGDP','Dai','Li','Maonan','Miao','She',
       'Zhuang','Ami_Stoneking','Atayal_Stoneking','Kinh_Vietnam_KHV']
myeas=["Longlin","LiangDao","Atayal_Stoneking","Japanese","Daur"]
neas=['HDYM1','Boisman_MN','Bbdong','Boshan','Linzi','Daur','Hezhen','Mongola','Ulchi','Xibo']
weas=['Tibetan_Shannan','Tibetan_Chamdo','Tibetan_Gangcha','Zongri', 'Chokhopani', 'Mebrak', 'Samdzong']

ty=["Tianyuan"]
ll=["Longlin"]
namer=["Mixe"]
#pD="/mnt/solexa/mel_yang/bronzeage/data/"
pD="/mnt/solexa/mel_yang/neolithic/qpwaveadm/"
fh="ind_A_v13_TBH5"

outs = [5]
comboswanted = ['00','01']
header = 'S1\tS2\tTarget\trank1\t%s_p\t%s_f1\t%s_f2\t%s_pnest\t%s_p\t%s_f1\t%s_f2\t%s_pnest\tse\n' % tuple(comboswanted[0:1]*4+comboswanted[1:2]*4)
rankswanted = ['1']
for outnum in outs:
    alldata = ""
    for ind1, i1 in enumerate(myeas):
        for ind2,i2 in enumerate(["HDYM1"]):
            for ind3,i3 in enumerate(neas+weas+seas):
                if i1==i2 or i2==i3 or i1==i3: continue
                s1,s2,target = (i1,i2,i3)
                #s1,s2,target = (i1,i2,i3)
                subset = 'X_%s_S1_%s_S2_%s_out%i' % (target,s1,s2,outnum)
                myinfo = pm.wave_extractAdmdata(pD,fh,subset,(s1,s2),target)
                if myinfo == 'NOFILE': continue              
                ranksprint = '\t'.join([str(myinfo['f4rank'][i]) for i in rankswanted])
                pvalsprint = '\t'.join(['%s\t%s\t%s\t%s' % tuple(myinfo['admix'][i]+myinfo['nestp'][i]) if i in myinfo['nestp'] else '%s\t%s\t%s\tNA' % tuple(myinfo['admix'][i]) for i in comboswanted])
                allprint = ranksprint+'\t'+pvalsprint+'\t'+myinfo['se'][-1]
                
                alldata+='\t'.join([s1,s2,target])+'\t'+allprint+'\n'
    #newname = 'adm_inSEASoutNWEASoutLL_out' + str(outnum)
    #newname = 'adm_inSEASoutNEASoutLD_out' + str(outnum)
    #newname = 'adm_inLLoutNEASoutLD_out' + str(outnum)
    #newname = 'adm_inLLoutWNSCEASoutTY_out' + str(outnum)
    #newname = 'adm_inLLoutSCEASoutDaur_out' + str(outnum)
    #newname = 'adm_inNSCEASoutNSCEASoutMixe_out' + str(outnum)
    #newname = 'adm_inLLoutSEASoutWNEAS_out' + str(outnum)
    newname = 'adm_inWNEASoutSEASoutSurui_out' + str(outnum)
    #newname = 'adm_inNCSEASoutLLoutLos_out' + str(outnum)
    newname = 'adm_inWNEASoutSEASoutLos_out' + str(outnum)
    newname = 'adm_inWNEASoutSEASoutChukchi_out' + str(outnum)
    newname = 'adm_inWNEASoutSEASoutHDYM1_out' + str(outnum)
    
    newfile = open(pD+fh+'.'+newname+'.tab','w')
    newfile.writelines(header+alldata)
    newfile.close()

```
