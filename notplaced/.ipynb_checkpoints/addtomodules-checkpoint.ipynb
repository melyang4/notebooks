```python
%%writefile /home/mel_yang/pylib/projmodules.py
#%%writefile /home/melinda_yang/pylib/projmodules.py

def convertf_par(inputfilename, outputfilename, outputformat, parfilename, indfilename=0,badsnpfile=0):
    ##inputfilename - w/ baseDir, thing to convert
    ##outputfilename - new fh
    ##outputformat, valid options include EIGENSTRAT ANCESTRYMAP PACKEDANCESTRYMAP etc
    ##parfilename, what to call this new par file
    ##indfilename, default is zero, no change, but if a new string, it searches for this
    ##string instead of inputfilename for the indfile.
    newfile = open(parfilename, 'w')
    newfile.write("genotypename:\t%s.geno\n" % inputfilename)
    newfile.write("snpname:\t%s.snp\n" % inputfilename)
    if indfilename == 0: newfile.write("indivname:\t%s.ind\n" % inputfilename)
    else: newfile.write("indivname:\t%s.ind\n" % indfilename)
    newfile.write("outputformat:\t%s\n" % outputformat)
    newfile.write("genotypeoutname:\t%s.geno\n" % outputfilename)
    newfile.write("snpoutname:\t%s.snp\n" % outputfilename)
    newfile.write("indivoutname:\t%s.ind\n" % outputfilename)
    if badsnpfile != 0: newfile.write("badsnpname:\t%s\n" % badsnpfile)
    newfile.close()
    return "Made par file for convertf!"

def Dparfile(pd,filehandle,label,f4mode=False,indfilename="None",badsnpfile=0,outpd=''):
    ##filehandle must not include suffix
    ##assumes all geno,snp,indiv files have the same filehandle and ends in
    ##.geno, .snp, and .ind
    ##Also assumes same for .pop file.
    ##outputs '.par' file ready to be run, assuming Dpopfile() was also called
    ##label==what to add to uniquify f3 par file
    ##f4mode=False; if set to true,then calculate f4 not D
    ##indfilename="None"; if set to a different filehandle than filehandle, 
    ##will use indfilename with that filehandle
    if f4mode == True: lab='f4'
    else:lab='D'
    if outpd=='': myfile = open(pd+filehandle+'.'+lab+'.'+label+'.par', 'w')
    else: myfile = open(outpd+filehandle+'.'+lab+'.'+label+'.par', 'w')
    myfile.write('genotypename:\t'+pd+filehandle+'.geno\n')
    myfile.write('snpname:\t'+pd+filehandle+'.snp\n')
    if indfilename=="None": myfile.write('indivname:\t'+pd+filehandle+'.ind\n')
    else: myfile.write('indivname:\t'+pd+indfilename+'.ind\n')
    if outpd=='': myfile.write('popfilename:\t'+pd+filehandle+'.'+lab+'.'+label+'.pop\n')
    else: myfile.write('popfilename:\t'+outpd+filehandle+'.'+lab+'.'+label+'.pop\n')
    myfile.write('printsd:\tYES\n')
    if badsnpfile != 0: myfile.write("badsnpname:\t%s\n" % badsnpfile)
    if f4mode == True: myfile.write('f4mode:\tYES\n')
    myfile.close()
    return "Made D par file" 

def Dpopfile(twopoplist,targetlist,ancestrallist,filehandle,label,f4mode=False):
    ##sourcelist must be list of tuples of two popns
    ##targetlist must be list of strings, either sourcelist admix into, or complete outgroup to sourcelist
    ##ancestrallist must be list of strings (some outgroup)
    ##filehandle does not include suffix
    ##label==what to add to uniquify f3 par file
    ##outputs '.pop' file, having matched every set of sources with each target
    if f4mode == True: lab='f4'
    else:lab='D'
    myfile = open(filehandle+'.'+lab+'.'+label+'.pop','w')
    count = 0
    for sources in twopoplist:
        for target in targetlist:
            for ancestral in ancestrallist:
                myfile.write(sources[0]+' '+sources[1]+' '+target+' '+ancestral+'\n')
                count+=1
    myfile.close()
    return count

def Dshfile(mynum,myinterval,filehandle,label,f4mode=False):
    mysteps = mynum/myinterval ##641/25
    if f4mode == True: lab='f4'
    else:lab='D'
    newfile = open(filehandle+'.'+lab+'.'+label+'.sh','w')
    parfile = filehandle+'.'+lab+'.'+label+'.par'
    logfile = filehandle+'.'+lab+'.'+label+'.log'
    newfile.write("rm %s\n" % (logfile))
    counter = 0
    while counter < mynum:
        newfile.write("qpDstat -p %s -l%s -h%s >> %s\n" % (parfile,str(counter+1),str(counter+myinterval),logfile))
        counter += myinterval
    newfile.close()
    return 'Made D sh file!'

def f3parfile(pD,fh,label,inbreed=False, indfh="", popfh="",outpd=''):
    ##filehandle must not include suffix
    ##assumes all geno,snp,indiv files have the same filehandle and ends in
    ##.geno, .snp, and .ind
    ##Also assumes same for .pop file.
    ##outputs '.par' file ready to be run, assuming f3popfile() was also called
    ##label==what to add to uniquify f3 par file
    ##inbreed==False, if True, add line inbreed: YES to parfile
    mypD=pD if outpd == "" else outpd
    myfile = open(mypD+fh+'.f3.'+label+'.par', 'w')
    myfile.write('genotypename:\t'+pD+fh+'.geno\n')
    myfile.write('snpname:\t'+pD+fh+'.snp\n')
    if indfh!="": myfile.write('indivname:\t'+pD+indfh+'.ind\n')
    else: myfile.write('indivname:\t'+pD+fh+'.ind\n')
    if popfh!="": myfile.write('popfilename:\t'+mypD+popfh+'.f3.'+label+'.pop\n')
    else: myfile.write('popfilename:\t'+mypD+fh+'.f3.'+label+'.pop\n')
    if inbreed == True: myfile.write('inbreed: YES')
    myfile.close()
    return "Made f3 par file"


def f3popfile(sourcelist,targetlist,filehandle,label):
    ##sourcelist must be list of tuples of two popns
    ##targetlist must be list of strings
    ##filehandle does not include suffix
    ##label==what to add to uniquify f3 par file
    ##outputs '.pop' file, having matched every set of sources with each target
    myfile = open(filehandle+'.f3.'+label+'.pop','w')
    for sources in sourcelist:
        for target in targetlist:
            myfile.write(sources[0]+' '+sources[1]+' '+target+'\n')
    myfile.close()
    return "Made f3 popn file"

def f4_extractinfo(projDirectory,Dfilelog,varyingindex):
    ##Get D statistic and Z score, sd and numSNPs considered
    ##varyingindex, list of IDs to note
    myfile = open(projDirectory+Dfilelog+'.log','r')
    myres = {}
    for line in myfile:
        x = line.strip().split()
        if x[0] != 'result:': continue
        D = float(x[-3])
        se = float(x[-2])
        Z = float(x[-1])
        label = tuple([x[i] for i in varyingindex])
        myres[label] = [D,Z,se]
    return myres

def f4parfile(pD,filehandle, subset, jackknife=0.01,indfh="",outpd=""):
    ##filehandle must not include suffix
    ##assumes all geno,snp,indiv files have the same filehandle and ends in
    ##.geno, .snp, and .ind
    ##Also assumes same for .pop file.
    ##outputs '.par' file ready to be run, assuming f4popfile() was also called
    ##jackknife (CM)=0.01, default is 1M bases=1 CM, if 10000 bases and blk is 0.01, then each block is 100 SNPs
    if outpd != "": myfile = open(outpd+filehandle+'.f4.'+subset+'.par', 'w')
    else: myfile = open(pD+filehandle+'.f4.'+subset+'.par', 'w')
    myfile.write('genotypename:\t'+pD+filehandle+'.geno\n')
    myfile.write('snpname:\t'+pD+filehandle+'.snp\n')
    if indfh != "": myfile.write('indivname:\t'+indfh+'.ind\n')
    else: myfile.write('indivname:\t'+pD+filehandle+'.ind\n')
    if outpd != "": myfile.write('popfilename:\t'+outpd+filehandle+'.f4.'+subset+'.pop\n')
    else: myfile.write('popfilename:\t'+pD+filehandle+'.f4.'+subset+'.pop\n')
    myfile.write('blgsize:\t'+str(jackknife)+'\n')
    myfile.close()
    return "Made F4 par file"


def f4popfile(sourcelist,targetlist,outgrouplist, filehandle,subset):
    ##sourcelist must be list of tuples of three popns (A,B,C)
    ##targetlist must be list of strings (X)
    ##outgrouplist must be list of strings (O)
    ##filehandle does not include suffix
    ##outputs '.pop' file, having matched every set of sources with each target
    myfile = open(filehandle+'.f4.'+subset+'.pop','w')
    for sources in sourcelist:
        for target in targetlist:
            for outgroup in outgrouplist:
                A,B,C,X,O = tuple(list(sources)+[target,outgroup])
                myfile.write("%s %s : %s %s:: %s %s : %s %s\n" % (A,O,X,C,A,O,B,C))
    myfile.close()
    return "Made F4 popn file"

def f4runitall(ABC, Xs, out, pD, fh, subset,indfh="",outpd=""):
    ##ABCs = sourcelist, list of tuples of three popns
    ##Xs = targetlist, list of strings
    ##out = outgrouplist, list of strings
    ##pD = project Directory
    ##fh = main filehandle
    ##subset = specific subset name to attach to filehandle, after '.f4.'
    from subprocess import Popen, PIPE
    if outpd!="":f4popfile(ABC, Xs, out, outpd+fh, subset)
    else: f4popfile(ABC, Xs, out, pD+fh, subset)
    f4parfile(pD,fh,subset,indfh=indfh,jackknife=0.01,outpd=outpd)
    if outpd!="": 
        outfile = open(outpd+fh+'.f4.'+subset+".log",'w')
        mycommand = 'qpF4ratio -p '+outpd+fh+'.f4.'+subset+'.par'
    else: 
        outfile = open(pD+fh+'.f4.'+subset+".log",'w')
        mycommand = 'qpF4ratio -p '+pD+fh+'.f4.'+subset+'.par'
    print 'Running command', mycommand
        
    varfile = Popen(mycommand.split(), stdout = outfile)
    varfile.communicate()
    outfile.close()
    return 'Finished running qp4Ratio'

def graph_par(pD,filehandle,subset,blgsize=0.05, outpop='Out',
                  lsqmode='YES',diag=0.0001,hires='YES',initmix=1000,
                  precision=0.0001,zthresh=3.0,terse='NO',useallsnps='NO',basegraph=False):
    ##filehandle must not include suffix
    ##assumes all geno,snp,indiv files have the same filehandle and ends in
    ##.geno, .snp, and .ind
    ##Also assumes same for .pop file.
    ##outputs '.par' file ready to be run, assuming f4popfile() was also called
    ##jackknife (CM)=0.01, default is 1M bases=1 CM, if 10000 bases and blk is 0.01, then each block is 100 SNPs
    if basegraph == False: parfile = open(pD+filehandle+'.graph.'+subset+'.par','w')
    else: parfile = open(pD+basegraph+'.'+filehandle+'.graph.'+subset+'.par','w')
    parfile.write('indivname:\t%s.ind\n' % (pD+filehandle))
    parfile.write('snpname:\t%s.snp\n' % (pD+filehandle))
    parfile.write('genotypename:\t%s.geno\n' % (pD+filehandle))
    parfile.write('outpop:\t%s\n' % outpop)
    parfile.write('blgsize:\t%f\n' % blgsize)
    parfile.write('lsqmode:\t%s\n' % lsqmode)
    parfile.write('diag:\t%f\n' % diag)
    parfile.write('hires:\t%s\n' % hires)
    parfile.write('initmix:\t%f\n' % initmix)
    parfile.write('precision:\t%f\n' % precision)
    parfile.write('zthresh:\t%f\n' % zthresh)
    parfile.write('terse:\t%s\n' % terse)
    parfile.write('useallsnps:\t%s\n' % useallsnps)
    parfile.close()
    return 'Made qpGraph par file!'


def graph_getidsig(pd,myname,zscorecutoff=3):
    import glob, os
    os.chdir(pd)
    id,zscore=([],[]) 
    for file in glob.glob(myname+'_'+"*.log"):
        #id.append(file.split(myname+'_')[1].split('.')[0])
        myfile = open(file,'r')
        for line in myfile: 
            if 'worst' not in line: continue
            id.append(file.split(myname+'_')[1].split('.')[0])
            zscore.append(line.split()[-1])
    mylst=[]
    for ind,i in enumerate(id): 
        if abs(float(zscore[ind]))<zscorecutoff: mylst.append((i,zscore[ind]))
    return mylst



def graph_retrieveinfo(sigdat,pd,subset,fh,myname):
    #print sigdat[0], sigdat[1]
    myid0 = myname.split('.')[0][:3].upper()+'0' 
    myid1 = myname.split('.')[0][:3].upper()+'1'
    myid2 = myname.split('.')[0][:3].upper()+'2'
    myfile = open(pd+myname+'_'+sigdat[0]+'.'+fh+'.'+subset+'.out','r')
    for line in myfile:
        if myid0 not in line and myid1 not in line and myid2 not in line: continue
        if 'vertex' in line: continue
        print line[:-1]
    myfile.close()
    return ''


def graph_runfile(basegraph,pD,filehandle,subset,blgsize=0.05,outpop='Out',
                  lsqmode='YES',diag=0.0001,hires='YES',initmix=1000,
                  precision=0.0001,zthresh=3.0,terse='NO',useallsnps='NO'):
    from subprocess import Popen, PIPE
    graph_par(pD,filehandle,subset,blgsize,outpop,lsqmode,diag,hires,
              initmix,precision,zthresh,terse,useallsnps)
    logfile = open(pD+basegraph+'.'+filehandle+'.'+subset+'.log','w')
    graphout = pD+basegraph+'.'+filehandle+'.'+subset
    mycommand = 'qpGraph -p '+pD+filehandle+'.graph.'+subset+'.par'
    mycommand += ' -g '+pD+basegraph+' -o '+graphout+'.out'
    mycommand += ' -d '+graphout+'.dot'
    print mycommand
    varfile = Popen(mycommand.split(), stdout = logfile)
    varfile.communicate()
    logfile.close()
    psfile = open(graphout+'.ps','w')
    pscommand = "dot -Tps "+graphout+".dot"
    varfile = Popen(pscommand.split(), stdout = psfile)
    varfile.communicate()
    psfile.close()
    pdfcommand = "ps2pdf "+graphout+".ps "+graphout+".pdf" 
    varfile = Popen(pdfcommand.split())
    varfile.communicate()
    return 'Running '+mycommand+' and making ps/pdf file'

def graph_poolrunfile(myargs):
    basegraph,pD,filehandle,subset=myargs
    blgsize=0.05
    outpop='Mbuti'
    lsqmode='YES'
    diag=0.0001
    hires='YES'
    initmix=1000
    precision=0.0001
    zthresh=3.0
    terse='NO'
    useallsnps='NO'
    from subprocess import Popen, PIPE
    graph_par(pD,filehandle,subset,blgsize,outpop,lsqmode,diag,hires,
              initmix,precision,zthresh,terse,useallsnps,basegraph)
    logfile = open(pD+basegraph+'.'+filehandle+'.'+subset+'.log','w')
    graphout = pD+basegraph+'.'+filehandle+'.'+subset
    mycommand = 'qpGraph -p '+pD+basegraph+'.'+filehandle+'.graph.'+subset+'.par'
    mycommand += ' -g '+pD+basegraph+' -o '+graphout+'.out'
    mycommand += ' -d '+graphout+'.dot'
    print mycommand
    varfile = Popen(mycommand.split(), stdout = logfile)
    varfile.communicate()
    logfile.close()
    psfile = open(graphout+'.ps','w')
    pscommand = "dot -Tps "+graphout+".dot"
    varfile = Popen(pscommand.split(), stdout = psfile)
    varfile.communicate()
    psfile.close()
    pdfcommand = "ps2pdf "+graphout+".ps "+graphout+".pdf" 
    varfile = Popen(pdfcommand.split())
    varfile.communicate()
    return 'Running '+mycommand+' and making ps/pdf file'

def graph_mkgraph(pD,filehandle,labels,edges,admixes,comments='NA',addadmix=False):
    ##labels = list of lists, listing [IDbefore,IDafter]
    ##edges = list of lists, listing [edgelabel,edgeinit,edgeend]
    ##admixes = list of lists, listing 
    ##[newadmixednode,1stpop,2ndpop,1stweight,2ndweight] 
    ##or [newadmixednode,1stpop,2ndpop] if don't know admix rate
    ##addadmix=False, True if do no admixweight, make sure to add weights!
    ##comments (optional), include som comments in beginning, some string
    graphfile = open(pD+filehandle,'w')
    if comments != 'NA': graphfile.write('##%s\n' % (comments))
    graphfile.write('root\tR\n')
    for label in labels: graphfile.write('label\t%s\t%s\n' % tuple(label[::-1]))
    for edge in edges: graphfile.write('edge\t%s\t%s\t%s\n' % tuple(edge))
    if len(admixes) != 0:
        for admix in admixes: 
            if len(admix)==3:
                graphfile.write('admix\t%s\t%s\t%s\n' % tuple(admix))
            elif len(admix)==5:
                admix1 = [float(i) if ind>2 else i for ind,i in enumerate(admix) ]
                graphfile.write('admix\t%s\t%s\t%s\t%f\t%f\n' % tuple(admix1))
            else: print 'weird'
    return 'Made basegraph'

def graph_readgraph(pD,filehandle):
    ##look in graphfile
    ##pull out all labels, edges, admixes
    #returns tuple of three lists (labels, edges, admixes)
    graphfile = open(pD+filehandle,'r')
    labels, edges, admixes = [],[],[]
    for line in graphfile:
        if line[0]=="#": continue
        x = line.strip().split()
        if x[0] == 'label': labels.append([x[2],x[1]])
        elif 'edge' in x[0]: edges.append(x[1:4])
        elif x[0] == 'admix': admixes.append(x[1:4])
        else: continue
    return labels,edges,admixes

def graph_addadmix(pD,fh,subset,basegraphname,wanttoadd,comments='NA',pool=0):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    from multiprocessing import Pool
    if pool != 0: myinputs=[]
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    num=0
    for ind1,myedges1 in enumerate(edgeset):
        for ind2,myedges2 in enumerate(edgeset[(ind1+1):]):
            labels,edges,admixes = labelset[:],edgeset[:],admixeset[:]
            ce1,ce2 = myedges1[0],myedges2[0]
            sublabel='_'+str(num)
            labels+=[wanttoadd[::-1]]
            dropped1 = [i for i in edges if ce1 in i][0]
            dropped2 = [i for i in edges if ce2 in i][0]
            newedges = [i for i in edges if ce1 not in i and ce2 not in i]
            newedges+=[[ce1+'1',dropped1[1],'I1'],
                       [ce2+'1',dropped2[1],'I2'],
                       [ce1+'2','I1',dropped1[2]],
                       [ce2+'2','I2',dropped2[2]],
                       [ce1+ce2+'a','I3',wanttoadd[0]]]
            admixes+=[['I3','I1','I2']]
            graph_mkgraph(pD,wanttoadd[0]+'.'+basegraphname+'_admix'+sublabel,labels,newedges,admixes,comments='NA')
            if pool == 0: graph_runfile(wanttoadd[0]+'.'+basegraphname+'_admix'+sublabel,pD,fh,subset,outpop='Mbuti')
            else:
                ##Add to myinputs
                myinputs.append((wanttoadd[0]+'.'+basegraphname+'_admix'+sublabel,pD,fh,subset))
            num+=1
    if pool !=0: 
        mypool = Pool(pool)
        results = mypool.map(graph_poolrunfile, myinputs)
    pdfout = pD+wanttoadd[0]+'.'+basegraphname+'_admix_ALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd[0]+'.'+basegraphname+'_admix*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs with admixture!"


def graph_addnode(pD,fh,subset,basegraphname,wanttoadd,comments='NA',pool=0):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    from multiprocessing import Pool
    
    if pool != 0: myinputs=[]
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    for ind,myedges in enumerate(edgeset):
        ce = myedges[0]
        sublabel='_'+str(ind)
        labels,edges,admixes=labelset[:],edgeset[:],admixeset[:]
        labels+=[wanttoadd[::-1]]
        dropped = [i for i in edges if ce in i][0]
        newedges = [i for i in edges if ce not in i]
        newedges+=[[ce+'1',dropped[1],'I1'],[ce+'2','I1',dropped[2]],[ce+'3','I1',wanttoadd[0]]]
        graph_mkgraph(pD,wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,labels,newedges,admixes,comments='NA')
        #graph_runfile(wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,pD,fh,subset,outpop='Mbuti')
        if pool == 0: graph_runfile(wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,pD,fh,subset,outpop='Mbuti')
        else:
            ##Add to myinputs
            myinputs.append((wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,pD,fh,subset))
        num+=1
    if pool !=0: 
        mypool = Pool(pool)
        results = mypool.map(graph_poolrunfile, myinputs)
    pdfout = pD+wanttoadd[0]+'.'+basegraphname+'_node_ALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd[0]+'.'+basegraphname+'_node*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs adding to each node!"



def graph_addadmix_uniq(pD,fh,subset,basegraphname,wanttoadd,comments='NA',pool=0):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    from multiprocessing import Pool
    
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    if pool != 0: myinputs=[]
    myid1,myid2,myid3 = tuple([wanttoadd[0].upper()[:3]+str(i) for i in range(3)])
    num=0
    for ind1,myedges1 in enumerate(edgeset):
        for ind2,myedges2 in enumerate(edgeset[(ind1+1):]):
            labels,edges,admixes = labelset[:],edgeset[:],admixeset[:]
            ce1,ce2 = myedges1[0],myedges2[0]
            sublabel=str(num)
            labels+=[wanttoadd[::-1]]
            
            dropped1 = [i for i in edges if ce1 in i][0]
            dropped2 = [i for i in edges if ce2 in i][0]
            newedges = [i for i in edges if ce1 not in i and ce2 not in i]
            newedges+=[[ce1+'1',dropped1[1],myid1],
                       [ce2+'1',dropped2[1],myid2],
                       [ce1+'2',myid1,dropped1[2]],
                       [ce2+'2',myid2,dropped2[2]],
                       [ce1+ce2+'a',myid3,wanttoadd[0]]]
            admixes+=[[myid3,myid1,myid2]]
            graph_mkgraph(pD,wanttoadd[0]+'.'+basegraphname+'_f'+sublabel,labels,newedges,admixes,comments='NA')
            #graph_runfile(wanttoadd[0]+'.'+basegraphname+'_f'+sublabel,pD,fh,subset,outpop='Mbuti')
            if pool == 0: graph_runfile(wanttoadd[0]+'.'+basegraphname+'_f'+sublabel,pD,fh,subset,outpop='Mbuti')
            else:
                ##Add to myinputs
                myinputs.append((wanttoadd[0]+'.'+basegraphname+'_f'+sublabel,pD,fh,subset))
            num+=1
    if pool !=0: 
        mypool = Pool(pool)
        results = mypool.map(graph_poolrunfile, myinputs)

    pdfout = pD+wanttoadd[0]+'.'+basegraphname+'_fALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd[0]+'.'+basegraphname+'_f*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs with admixture!"


def graph_addnode_uniq(pD,fh,subset,basegraphname,wanttoadd,comments='NA',pool=0):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    from multiprocessing import Pool
    
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    if pool != 0: myinputs=[]
    myid1,myid2,myid3 = tuple([wanttoadd[0].upper()[:3]+str(i) for i in range(3)])
    for ind,myedges in enumerate(edgeset):
        ce = myedges[0]
        sublabel=str(ind)
        labels,edges,admixes=labelset[:],edgeset[:],admixeset[:]
        labels+=[wanttoadd[::-1]]
        dropped = [i for i in edges if ce in i][0]
        newedges = [i for i in edges if ce not in i]
        newedges+=[[ce+'1',dropped[1],myid1],[ce+'2',myid1,dropped[2]],[ce+'3',myid1,wanttoadd[0]]]
        graph_mkgraph(pD,wanttoadd[0]+'.'+basegraphname+'_n'+sublabel,labels,newedges,admixes,comments='NA')
        #graph_runfile(wanttoadd[0]+'.'+basegraphname+'_n'+sublabel,pD,fh,subset,outpop='Mbuti')
        if pool == 0: graph_runfile(wanttoadd[0]+'.'+basegraphname+'_n'+sublabel,pD,fh,subset,outpop='Mbuti')
        else:
            ##Add to myinputs
            myinputs.append((wanttoadd[0]+'.'+basegraphname+'_n'+sublabel,pD,fh,subset))
    if pool !=0: 
        mypool = Pool(pool)
        results = mypool.map(graph_poolrunfile, myinputs)
    pdfout = pD+wanttoadd[0]+'.'+basegraphname+'_nALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd[0]+'.'+basegraphname+'_n*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs adding to each node!"

def graph_substitute_uniq(pD,fh,subset,basegraphname,original,newpop,comments='NA'):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    
    omyid1,omyid2,omyid3 = tuple([original.upper()[:3]+str(i) for i in range(3)])
    nmyid1,nmyid2,nmyid3 = tuple([newpop.upper()[:3]+str(i) for i in range(3)])
    oldset = [original, omyid1, omyid2, omyid3]
    newset = [newpop, nmyid1, nmyid2, nmyid3]
    
    myoldgraph = open(pD+basegraphname,'r')
    newgraphlines = []
    for line in myoldgraph:
        newline = line.replace(oldset[0],newset[0]).replace(oldset[1],newset[1]).replace(oldset[2],newset[2]).replace(oldset[3],newset[3])
        newgraphlines.append(newline)
    myoldgraph.close()
    newbasegraphname = basegraphname.replace(original,'sub'+newpop)
    mynewgraph = open(pD+newbasegraphname,'w')
    mynewgraph.writelines(newgraphlines)
    mynewgraph.close()
    
    graph_runfile(newbasegraphname,pD,fh,subset,outpop='Mbuti')
    
    return "Substituted a label!!"


def graph_addadmix_two(pD,fh,subset,basegraphname,wanttoadd,wanttoadd_B,comments='NA'):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    for ind,i in enumerate(labelset):
        if 'NAMER' in i: labelset[ind]=wanttoadd_B
    for ind,i in enumerate(edgeset):
        if 'NAMER' in i: edgeset[ind]=['xmf0xa','f1I3',wanttoadd_B[1]]

    num=0
    for ind1,myedges1 in enumerate(edgeset):
        for ind2,myedges2 in enumerate(edgeset[(ind1+1):]):
            labels,edges,admixes = labelset[:],edgeset[:],admixeset[:]
            ce1,ce2 = myedges1[0],myedges2[0]
            sublabel='_'+str(num)
            labels+=[wanttoadd]
            dropped1 = [i for i in edges if ce1 in i][0]
            dropped2 = [i for i in edges if ce2 in i][0]
            newedges = [i for i in edges if ce1 not in i and ce2 not in i]
            newedges+=[[ce1+'1',dropped1[1],'I1'],
                       [ce2+'1',dropped2[1],'I2'],
                       [ce1+'2','I1',dropped1[2]],
                       [ce2+'2','I2',dropped2[2]],
                       [ce1+ce2+'a','I3',wanttoadd[1]]]
            admixes+=[['I3','I1','I2']]
            graph_mkgraph(pD,wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_admix'+sublabel,labels,newedges,admixes,comments='NA')
            graph_runfile(wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_admix'+sublabel,pD,fh,subset,outpop='Mbuti')
            num+=1
    pdfout = pD+wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_admix_ALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_admix*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs with admixture!"


def graph_addnode_two(pD,fh,subset,basegraphname,wanttoadd,wanttoadd_B,comments='NA'):
    from subprocess import Popen, PIPE
    import projmodules as pm
    import glob
    labelset,edgeset,admixeset=pm.graph_readgraph(pD,basegraphname)
    
    for ind,i in enumerate(labelset):
        if 'NAMER' in i: labelset[ind]=wanttoadd_B
    for ind,i in enumerate(edgeset):
        if 'NAMER' in i: edgeset[ind]=['xmf0xa','f1I3',wanttoadd_B[1]]
    
    
    for ind,myedges in enumerate(edgeset):
        ce = myedges[0]
        sublabel='_'+str(ind)
        labels,edges,admixes=labelset[:],edgeset[:],admixeset[:]
        labels+=[wanttoadd]
        dropped = [i for i in edges if ce in i][0]
        newedges = [i for i in edges if ce not in i]
        newedges+=[[ce+'1',dropped[1],'I1'],[ce+'2','I1',dropped[2]],[ce+'3','I1',wanttoadd[1]]]
        graph_mkgraph(pD,wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,labels,newedges,admixes,comments='NA')
        graph_runfile(wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_node'+sublabel,pD,fh,subset,outpop='Mbuti')
    pdfout = pD+wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_node_ALL.'+fh+'.'+subset+".pdf"
    varfile = Popen(['rm',pdfout])
    varfile.communicate()
    files = glob.glob(pD+wanttoadd_B[0]+'.'+wanttoadd[0]+'.'+basegraphname+'_node*'+fh+'.'+subset+".pdf")
    varfile = Popen(['pdftk']+files+['cat','output']+[pdfout])
    varfile.communicate()
    varfile = Popen(['pdfcrop']+[pdfout,pdfout])
    varfile.communicate()
    varfile = Popen(['rm']+files)
    varfile.communicate()
    return "Made new graphs adding to each node!"



def indtoignore(listtonotignore, inputfile, outputfile, firstcolcharlim=20):
    ##listtonotignore--popn names to leave as is, all others change to Ignore
    ##inputfile
    ##outputfile
    ##firstcolcharlim, default twenty spaces, add blanks in front of first column as needed
    indfile = open(inputfile,'r')
    newfile = open(outputfile,'w')
    for ind,line in enumerate(indfile):
        x = line.split()
        x[0] = ' '*(firstcolcharlim-len(x[0])) + x[0]
        if x[2] not in listtonotignore: 
            x[2] = "Ignore"
            newfile.write('%s %s %s\n' % tuple(x))
        else: 
            newfile.write(line)  
    indfile.close()
    newfile.close()
    return 'Made ind file for convertf, done!'

def indtoignore_spec(listtonotignore, labelnames, inputfile, outputfile, firstcolcharlim=20):
    ##listtonotignore--popn names to change to labelnames list, all others change to Ignore
    ##labelnames = list of title to give, if NA then keep same
    ##inputfile
    ##outputfile
    ##firstcolcharlim, default twenty spaces, add blanks in front of first column as needed
    indfile = open(inputfile,'r')
    newfile = open(outputfile,'w')
    for ind,line in enumerate(indfile):
        x = line.split()
        x[0] = ' '*(firstcolcharlim-len(x[0])) + x[0]
        if x[2] not in listtonotignore: 
            x[2] = "Ignore"
            newfile.write('%s %s %s\n' % tuple(x))
        else: 
            myindex = listtonotignore.index(x[2])
            if labelnames[myindex] == 'NA': newfile.write(line)  
            else: x[2] = labelnames[myindex]; newfile.write('%s %s %s\n' % tuple(x))
    indfile.close()
    newfile.close()
    return 'Made ind file for convertf, done!'


def makeDtargetlist_varyp1_p2(groups,labels,p3,p4):
    ##groups = tuple of 1 or 2 lists
    ##labels = tuple of labelnames corresponding to number of groups
    ##p3,p4 = labels for p3 and p4 in D
    if len(groups) == 1:
        mytuples = [(i,j,p3,p4,labels[0]) for ind,i in enumerate(groups[0]) for j in groups[0][(ind+1):]]
        return mytuples
    elif len(groups) == 2:
        within0 = [(i,j,p3,p4,labels[0]) for ind,i in enumerate(groups[0]) for j in groups[0][(ind+1):]]
        between = [(i,j,p3,p4,'%s_v_%s' % (labels[0],labels[1])) for i in groups[0] for j in groups[1]]
        within1 = [(i,j,p3,p4,labels[1]) for ind,i in enumerate(groups[1]) for j in groups[1][(ind+1):]]
        mytuples = within0+between+within1
        return mytuples
    else: return "Too many groups!"
    
def makeDtargetlist_varyp1_p3(groups,labels,p2,p4):
    ##groups = tuple of 1 or 2 lists
    ##labels = tuple of labelnames corresponding to number of groups
    ##p2,p4 = labels for p2 and p4 in D
    if len(groups) == 1:
        mytuples = [(i,p2,j,p4,labels[0]) for ind,i in enumerate(groups[0]) for j in groups[0][(ind+1):]]
        return mytuples
    elif len(groups) == 2:
        within0 = [(i,p2,j,p4,labels[0]) for ind,i in enumerate(groups[0]) for j in groups[0][(ind+1):]]
        between = [(i,p2,j,p4,'%s_v_%s' % (labels[0],labels[1])) for i in groups[0] for j in groups[1]]
        within1 = [(i,p2,j,p4,labels[1]) for ind,i in enumerate(groups[1]) for j in groups[1][(ind+1):]]
        mytuples = within0+between+within1
        return mytuples
    else: return "Too many groups!"

def projection_eigen(filehandle, reference, test, ancestor, bin=0,testind='random'):
    ##filehandle does not include suffix
    ##needs to have '.geno', '.ind', '.snp' files attached to filehandle
    ##reference, test and ancestor are popn ids in the .ind file, third column
    ##bin default is 0, meaning no bins, otherwise
    ##specify a bin and it will bin by frequency
    ##outputs projections for each test, relative to the reference DAF
    ##output is in numpy array format, ready for np.loadtxt
    ##each row is an individual, each column is a DAF for the reference
    import numpy as np
    mygenofilename = filehandle+".geno"
    myindfilename = filehandle+".ind"
    mysnpfilename = filehandle+".snp"
    
    ##Get reference and test indices
    referenceinds = []
    testinds = []
    ancinds = []
    myindfile = open(myindfilename,'r')
    for ind,line in enumerate(myindfile): 
        x = line.split()
        if x[2] == reference: referenceinds.append(ind)
        if x[2] == test: testinds.append(ind)
        if x[2] == ancestor: ancinds.append(ind)
    if testind != 'random' and len(testind) < len(testinds): testinds = [testinds[testind]]
    if len(referenceinds) == 0 or len(testinds) == 0 or len(ancinds) == 0: 
        return "One of sets empty!", referenceinds, testinds, ancinds
    print "Found test/ref/anc indices"

    if bin == 0: 
        if testind=='random': proj = np.array([[0 for i in range((len(referenceinds)*2)-1)]])
        else: proj = np.array([[0 for i in range((len(referenceinds)*2)-1)] for j in testinds])
        #proj = np.array([[0 for i in range((len(referenceinds)*2)-1)] for j in testinds])
        totalpersite = np.array([0 for i in range((len(referenceinds)*2)-1)])
    else: 
        if testind=='random': proj =np.array([[0 for i in range(bin-1)]])
        else: proj = np.array([[0 for i in range(bin-1)] for j in testinds])
        totalpersite = np.array([0 for i in range(bin-1)])
    print proj
    myfile = open(mygenofilename,'r')

    for ind,line in enumerate(myfile): 
        mygenotypes = line.rstrip()
        ref = np.array([int(mygenotypes[i]) for i in referenceinds])
        tes = np.array([int(mygenotypes[i]) for i in testinds])
        anc = np.array([int(mygenotypes[i]) for i in ancinds])
        
        ##Ancestor must be fixed
        anc = list(set(anc))
        if len(anc) != 1: continue
        if anc[0] not in [0,2]: continue
        anc = anc[0]
        
        ##Adjust for ancestry if needed
        if anc == 2: 
            ref = 2-ref
            tes = 2-tes
        else: pass
        
        if sorted(list(set(ref))) not in [[0],[0,1],[0,1,2],[1],[1,2],[2]]: continue
        if testind != 'random':
            if sorted(list(set(tes))) not in [[0],[0,1],[0,1,2],[1],[1,2],[2],[0,2]]: continue
        if testind == 'random':
            import random
            if sorted(list(set(tes))) in [[-7],[9],[-7,9]]: continue
            tes = np.array([random.choice([i for i in tes if i in [0,1,2]])])
        #print tes    
        totalref, refind = ref.shape[0]*2.0, np.sum(ref,axis=0)
        reffreq = refind/totalref
        if bin != 0: refind = int(bin*reffreq)
        if refind in [0, totalref]: continue
        totalpersite[refind-1] += 2.0
        for ind1,mygeno in enumerate(tes): proj[ind1][refind-1] += mygeno
    print proj
    expected = totalpersite*((np.array(range(1,int(totalref))))/totalref)

    for indiv in range(proj.shape[0]):
        if indiv == 0: adjustproj = proj[indiv,]/expected
        else: adjustproj = np.vstack((adjustproj,proj[indiv,]/expected))
        ##adjustproj.append([i/expected[ind2] for ind2,i in enumerate(indiv)])
    if bin == 0: np.savetxt(filehandle+'.r'+reference+'.t'+test+'.a'+ancestor+'.proj', adjustproj, fmt='%.5f', delimiter='\t')
    else: np.savetxt(filehandle+'.b'+str(bin)+'.r'+reference+'.t'+test+'.a'+ancestor+'.proj', adjustproj, fmt='%.5f', delimiter='\t')
    return 'Done '+reference+' '+test+' '+ancestor+' combo!'


def smartpca_par(pd,fh, datpd="", outlier=0, xregionname="NA",lsqproject=False,fstonly=False,
                 newfilename="NA",indfilename="NA",poplistname="NA"):
    ##fh - filehandle of data
    
    inputfilename=pd+fh if datpd=="" else datpd+fh
    
    newfile = open(pd+fh+".smartpca.par", 'w')
    newfile.write("genotypename:\t%s.geno\n" % inputfilename)
    newfile.write("snpname:\t%s.snp\n" % inputfilename)
    if indfilename == "NA": newfile.write("indivname:\t%s.ind\n" % inputfilename)
    else: newfile.write("indivname:\t%s.ind\n" % indfilename)
    if newfilename=="NA":
        newfile.write("evecoutname:\t%s.evec\n" % pd+fh)
        newfile.write("evaloutname:\t%s.eval\n" % pd+fh)
    else:
        newfile.write("evecoutname:\t%s.evec\n" % pd+fh)
        newfile.write("evaloutname:\t%s.eval\n" % pd+fh)
    if outlier != 0: newfile.write("outliermode:\t%s\n" % str(outlier))
    if xregionname != "NA": newfile.write("xregionname:\t%s\n" % str(xregionname))
    if lsqproject==True: newfile.write("lsqproject:\tYES\n")
    if poplistname != "NA": newfile.write("poplistname:\t%s\n" % str(poplistname))
    if fstonly==True: newfile.write("fstonly:\tYES\n")
    newfile.close()
    return "Made par file for smartpca!"

def smartpca_par_old(inputfilename, outlier=0, xregionname="NA",lsqproject=False,fstonly=False,
                 newfilename="NA",indfilename="NA",poplistname="NA"):
    ##inputfilename - w/ baseDir, thing to convert
    newfile = open(inputfilename+".smartpca.par", 'w')
    newfile.write("genotypename:\t%s.geno\n" % inputfilename)
    newfile.write("snpname:\t%s.snp\n" % inputfilename)
    if indfilename == "NA": newfile.write("indivname:\t%s.ind\n" % inputfilename)
    else: newfile.write("indivname:\t%s.ind\n" % indfilename)
    if newfilename=="NA":
        newfile.write("evecoutname:\t%s.evec\n" % inputfilename)
        newfile.write("evaloutname:\t%s.eval\n" % inputfilename)
    else:
        newfile.write("evecoutname:\t%s.evec\n" % newfilename)
        newfile.write("evaloutname:\t%s.eval\n" % newfilename)
    if outlier != 0: newfile.write("outliermode:\t%s\n" % str(outlier))
    if xregionname != "NA": newfile.write("xregionname:\t%s\n" % str(xregionname))
    if lsqproject==True: newfile.write("lsqproject:\tYES\n")
    if poplistname != "NA": newfile.write("poplistname:\t%s\n" % str(poplistname))
    if fstonly==True: newfile.write("fstonly:\tYES\n")
    newfile.close()
    return "Made par file for smartpca!"


def treemix_mkinput(pd,myfilename,subset,popnsiwant,indfilename="NA",includemissing=False):
    ##pd=project Directory
    ##myfilename==fh before the '.ind' (needs to be from an eigenstrat format
    ##subset==name to add on (and final name of files)
    ##popnsiwant==list of popns to include in treemix file
    ##if indfilename!="NA" then it needs to specify name of indfile, 
    ##which can be different from geno/snp file (but make sure it lines up correctly)
    ##Currently discards site if any missing data
    import numpy as np
    import random
    
    if indfilename == "NA": myindfile = open(pd+myfilename+'.ind', 'r')
    else: myindfile = open(pd+indfilename+'.ind', 'r')
    popindex=[[] for i in popnsiwant]
    for ind,line in enumerate(myindfile):
        mypop = line.strip().split()[2]
        if mypop in popnsiwant: 
            popindex[popnsiwant.index(mypop)].append(ind)
    myindfile.close()

    treemixindfile = open(pd+subset+'.popns.txt','w')
    treemixindfile.write('\n'.join(popnsiwant)+'\n')
    treemixindfile.close()
    
    myfile = open(pd+myfilename+'.geno','r')
    
    newfile = open(pd+myfilename+'.'+subset+'.treemix.txt','w')
    newfile.write(' '.join(popnsiwant)+'\n')
    
    if includemissing==False:
        popindexmerge = [item for sublist in popindex for item in sublist]
        num=0
        for ind,line in enumerate(myfile):
            tempary = np.array([int(i) for i in line.strip()])
            tempary1 = tempary[popindexmerge]
            if 9 in tempary1: continue
            newline = ''
            for ind2,myset in enumerate(popindex):
                myarys=tempary[myset]
                snp1 = np.sum(myarys)
                snp2 = (2*myarys.shape[0])-snp1
                newline += '%d,%d ' % (snp1,snp2)
            newfile.write( newline[:-1] + '\n')
            ##if num == 1: break
            num+=1
    if includemissing==True:
        for ind,line in enumerate(myfile):
            #if ind>5: break 
            newline=''
            for ind2,myset in enumerate(popindex):
                mygenos=sum([int(line[pos]) for pos in myset if line[pos]!='9'])
                mycounts=sum([2 for pos in myset if line[pos]!='9'])
                #print popnsiwant[ind2], mygenos, mycounts, [line[pos] for pos in myset]
                if mycounts==0: newline='pleasecontinue'; break
                else: newline += '%d,%d ' % (mygenos,mycounts-mygenos)
                #elif mycounts==2: 
                #    if mygenos==0: newline += '%d,%d ' % (0,2)
                #    elif mygenos==1: newline += '%d,%d ' % (1,1)
                #    elif mygenos==2: newline += '%d,%d ' % (2,0)
                #    else: print mygenos, 'wha???'
                #else: 
                #    fullset=['a']*mygenos+['b']*(mycounts-mygenos)
                #    sampledalleles=''.join(random.sample(fullset,2))
                #    if sampledalleles == 'aa': newline += '%d,%d ' % (2,0)
                #    elif sampledalleles == 'bb': newline += '%d,%d ' % (0,2)
                #    elif sampledalleles in ['ab','ba']: newline += '%d,%d ' % (1,1)
                #    else: print 'whaaa?', sampledalleles
                
            if newline=='pleasecontinue': continue
            #print newline
            newfile.write( newline[:-1] + '\n')
               
    newfile.close()
    myfile.close()
    return "Made Treemix file! GZip it!"

def treemix_mkinput_old(pd,myfilename,subset,popnsiwant,indfilename="NA"):
    ##pd=project Directory
    ##myfilename==fh before the '.ind' (needs to be from an eigenstrat format
    ##subset==name to add on (and final name of files)
    ##popnsiwant==list of popns to include in treemix file
    ##if indfilename!="NA" then it needs to specify name of indfile, 
    ##which can be different from geno/snp file (but make sure it lines up correctly)
    ##Currently discards site if any missing data
    import numpy as np
    
    if indfilename == "NA": myindfile = open(pd+myfilename+'.ind', 'r')
    else: myindfile = open(pd+indfilename+'.ind', 'r')
    popindex=[[] for i in popnsiwant]
    for ind,line in enumerate(myindfile):
        mypop = line.strip().split()[2]
        if mypop in popnsiwant: 
            popindex[popnsiwant.index(mypop)].append(ind)
    myindfile.close()
    popindexmerge = [item for sublist in popindex for item in sublist]
    
    treemixindfile = open(pd+subset+'.popns.txt','w')
    treemixindfile.write('\n'.join(popnsiwant)+'\n')
    treemixindfile.close()
    
    myfile = open(pd+myfilename+'.geno','r')
    
    newfile = open(pd+myfilename+'.'+subset+'.treemix.txt','w')
    newfile.write(' '.join(popnsiwant)+'\n')
    num=0
    for ind,line in enumerate(myfile):
        tempary = np.array([int(i) for i in line.strip()])
        tempary1 = tempary[popindexmerge]
        if 9 in tempary1: continue
        newline = ''
        for ind2,myset in enumerate(popindex):
            myarys=tempary[myset]
            snp1 = np.sum(myarys)
            snp2 = (2*myarys.shape[0])-snp1
            newline += '%d,%d ' % (snp1,snp2)
        newfile.write( newline[:-1] + '\n')
        ##if num == 1: break
        num+=1
    newfile.close()
    myfile.close()
    return "Made Treemix file! GZip it!"

def wave_par(mypD, fh, subset,indfilename="None",badsnpfile=0,outpd=''):
    ##pD = project Directory
    ##fh = main filehandle
    ##subset = specific subset name to attach to filehandle, after '.wave.'
    pD = mypD if outpd=='' else outpd
    parfile = open(pD+fh+'.wave.'+subset+'.par','w')
    parfile.write('genotypename:\t%s.geno\n' % (mypD+fh))
    if badsnpfile != 0: parfile.write('snpname:\t%s\n' % (mypD+badsnpfile))
    else: parfile.write('snpname:\t%s.snp\n' % (mypD+fh))
    if indfilename != "None": parfile.write('indivname:\t%s\n' % (mypD+indfilename))
    else: parfile.write('indivname:\t%s.ind\n' % (mypD+fh))
    parfile.write('popleft:\t%s.wave.%s.leftpop\n' % (pD+fh,subset))
    parfile.write('popright:\t%s.wave.%s.rightpop\n' % (pD+fh,subset))
    parfile.close()
    return 'Made qpWave par file!'
    
def wave_pop(poplist, leftorright, pD, fh, subset):
    ##poplist -- list of popns to put into file, either left of right population
    ##leftorright -- is this popn list for the left or right pop file? Type 'left' or 'right'
    ##pD = project Directory
    ##fh = main filehandle
    ##subset = specific subset name to attach to filehandle, after '.wave.'
    popfile = open(pD+fh+'.wave.'+subset+'.'+leftorright+'pop','w')
    for i in poplist: popfile.write(i+'\n')
    popfile.close()
    return 'Made qpWave pop files!'

def wave_run(testpoplist, outpoplist, mypD, fh, subset,indfilename="None",badsnpfile=0,outpd=''):
    ##testpoplist -- list of popns to put into file, in left
    ##outpoplist -- list of popns to put into file, in right
    ##pD = project Directory
    ##fh = main filehandle
    ##subset = specific subset name to attach to filehandle, after '.wave.'
    from subprocess import Popen, PIPE
    pD = mypD if outpd=='' else outpd
    wave_pop(testpoplist, 'left', pD, fh, subset)
    wave_pop(outpoplist, 'right', pD, fh, subset)
    wave_par(mypD, fh, subset,indfilename,badsnpfile,outpd)
    outfile = open(pD+fh+'.wave.'+subset+".log",'w')
    mycommand = 'qpWave -p '+pD+fh+'.wave.'+subset+'.par'
    print 'Running command', mycommand
        
    varfile = Popen(mycommand.split(), stdout = outfile)
    varfile.communicate()
    outfile.close()
    return 'Finished running qpWave'

def wave_adm_run(testpoplist, outpoplist, mypD, fh, subset,indfilename="None",badsnpfile=0,outpd=''):
    ##testpoplist -- list of popns to put into file, in left
    ##outpoplist -- list of popns to put into file, in right
    ##pD = project Directory
    ##fh = main filehandle
    ##subset = specific subset name to attach to filehandle, after '.wave.'
    ##requires having run wave_run beforehand, uses wave_run's par file
    from subprocess import Popen, PIPE
    pD = mypD if outpd=='' else outpd
    wave_pop(testpoplist, 'left', pD, fh, subset)
    wave_pop(outpoplist, 'right', pD, fh, subset)
    wave_par(mypD, fh, subset,indfilename,badsnpfile,outpd)
    outfile = open(pD+fh+'.adm.'+subset+".log",'w')
    mycommand = 'qpAdm -p '+pD+fh+'.wave.'+subset+'.par'
    print 'Running command', mycommand
        
    varfile = Popen(mycommand.split(), stdout = outfile)
    varfile.communicate()
    outfile.close()
    return 'Finished running qpAdm'

def wave_parselogfile(logfilename):
    ##logfilename: name of logfile where i should 
    ##grab info--self made, see above
    logfile = open(logfilename,'r')
    mydata = {}
    for ind,line in enumerate(logfile):
        if '[' in line:
            x1 = tuple(line[1:-2].replace("'",'').replace(' ','').split(','))
            mydata[x1] = []
        elif 'f4rank' in line:
            x2 = line.split()
            tail = float(x2[7])
            mydata[x1].append(tail)
    logfile.close()
    return mydata

def wave_extractAdmdata(pD,fh,subset,sources,target):
    admfile = open(pD+fh+'.adm.'+subset+'.log','r')
    admlines = admfile.readlines()
    rankindex = [line.split() for ind,line in enumerate(admlines) if 'f4rank' in line]
    f4ranks = [(entry[1],float(entry[7])) for entry in rankindex]
    fixedindex = [ind for ind,line in enumerate(admlines) if 'fixed' in line]
    #print [line.split() for line in admlines if 'std' in line]
    stderror = [line.split() for line in admlines if 'std' in line][0][2:]
    
    if len(sources) == 2: myindex = 6
    if len(fixedindex) == 0: return 'NOFILE'
    reldata = [line.split() for line in admlines[fixedindex[0]:(fixedindex[0]+myindex)]]
    
    extdat = [entry[0:1]+entry[4:] for entry in reldata if entry[0][0] in '01']
    nested = [entry[2:3]+[entry[-1]] for entry in reldata if entry[0][0] == 'b']
    f4rankdict, extdict, nestp = {},{},{}
    for i in f4ranks: f4rankdict[i[0]] = i[1]
    for i in extdat: 
        if 'infeasible' in i:
            numNA = len(i[1:-1])
            extdict[i[0]] = ['NA']*numNA
        else:
            extdict[i[0]] = i[1:]
    for i in nested: nestp[i[0]] = [i[1]]
    info = {'f4rank':f4rankdict,'admix':extdict,'nestp':nestp,'se':stderror}
    return info
    
def wave_makeary(logdata, pD, fh, twolists='NA', transpose='Yes', rank=0):
    ##logfilename==logfile to parse
    ##twolists=='NA', assume list is made up one from logdata tuples
    ##otherwise, must be list/tuple of two lists, to be indivs for ary
    ##transpose=='Yes', transpose array, otherwise don't transpose
    import numpy as np
    from collections import OrderedDict as od
    if twolists == 'NA':
        myids1 = list(od.fromkeys([i[0] for i in logdata.keys()]))
        myids2 = list(od.fromkeys([i[1] for i in logdata.keys()]))
    else: myids1, myids2 = tuple(twolists)
    myary = np.zeros((len(myids1),len(myids2)))
    for mypair in logdata:
        if mypair[0] in myids1 and mypair[1] in myids2:
            myary[myids1.index(mypair[0]),myids2.index(mypair[1])] = logdata[mypair][rank]
        if mypair[1] in myids1 and mypair[0] in myids2:
            myary[myids1.index(mypair[1]),myids2.index(mypair[0])] = logdata[mypair][rank]
    if twolists != 'NA' and twolists[0] == twolists[1]:
        for i in range(myary.shape[0]): myary[i,i] = 'nan'  
    aryfile = open(pD+fh+'.tab','w')
    if transpose == 'Yes': 
        myary = myary.transpose()
        aryfile.write('Pops\t'+'\t'.join([i for i in myids1])+'\n')
        for i in range(myary.shape[0]):
            aryfile.write(myids2[i]+'\t'+'\t'.join([str(i) for i in myary[i,]])+'\n')
    else: 
        aryfile.write('Pops\t'+'\t'.join([i for i in myids2])+'\n')
        for i in range(myary.shape[0]):
            aryfile.write(myids1[i]+'\t'+'\t'.join([str(i) for i in myary[i,]])+'\n')
    aryfile.close()
    return 'Made array for qpWave '+pD+fh+'.tab'
    
```
