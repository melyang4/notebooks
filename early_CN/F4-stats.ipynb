```python
import matplotlib.pyplot as plt
import matplotlib

myinds=["LiangDao","Suogang","Qihe","Jiahu","Baojianshan","Tianluoshan",
        "Bbdong","Boshan","Donghulin","Zhaoguo","XJS","JomonM"]
fig = plt.figure(figsize=(20,10))
rowcol=mylst


for mycount,myind in enumerate(myinds):
    ax=fig.add_subplot(3,4,mycount+1)
    for xind,mypop in enumerate(rowcol):
        if xind==rowcol.index(myind): continue
        ax.plot(myary[rowcol.index(myind),xind],myary[xind,rowcol.index(myind)],
                marker=lowhighdict[mypop][2], c=lowhighdict[mypop][1],
                label=mypop,linestyle="None")
    ax.set_title(myind)
    ax.axvline(x=0, ls='solid', c='gray')
    ax.axhline(y=0, ls='solid', c='gray')
plt.tight_layout()
ax.legend(loc=4,ncol=11,fontsize=8,bbox_to_anchor=(0.5, -0.5))
plt.show()
```
