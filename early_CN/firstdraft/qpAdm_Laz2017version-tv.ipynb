````python
%%writefile /mnt/solexa/mel_yang/early_CN/firstdraft/qpadm_Laz2017/generalSStv.py
import fstats_funcs as ff
import subprocess
import sys

"""
python generalSStv.py 15 Qihe o1 Qihe
"""

name="SHK"
out="o1"
target="Shamanka_EN"
pool,name,out,target=tuple(sys.argv[1:])

pD="/mnt/solexa/mel_yang/data/2.2M/" 
outpd0="/mnt/solexa/mel_yang/early_CN/firstdraft/qpadm_Laz2017/"
outpd=outpd0+name+"/"
fh="UPA19_3"
pt="adm"

basefile=open(outpd0+fh+"."+out,'r')
base=[line.strip() for line in basefile]
basefile.close()

myinputs=[]
for ind1,s1 in enumerate(base):
    newbase=base[:]
    newbase.remove(s1)
    label="%s_SStv_%s_%s_%s" % (name,target,s1,out)
    ff.parfile(pD,fh,pt,label,outpd=outpd,outgroup=fh+"."+label,indfh='UPA19_3.qpadmmerge1',badsnp=fh+".transi.snp\n")
    ff.mklst(outpd,fh,label,newbase)
    ff.mklst(outpd,fh,pt+"."+label,[target,s1],suffix="leftpop")
    myinputs.append(outpd+fh+".adm."+label)

print len(myinputs)
    
def runqpadm(filehandle):
    from subprocess import Popen, PIPE
    parfilename=filehandle+".par"
    logfilename=filehandle+".log"
    logfile = open(logfilename,'w')
    mycommand="qpAdm -p "+parfilename
    print mycommand
    varfile = Popen(mycommand.split(), stdout = logfile)
    varfile.communicate()
    logfile.close()
    return 'Running '+mycommand

from multiprocessing import Pool
mypool = Pool(int(pool))
results = mypool.map(runqpadm, myinputs)
````
