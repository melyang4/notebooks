````python
import matplotlib.pyplot as plt
import matplotlib

haak=['Karelia_HG','Motala_HG','LBK_EN','Unetice_EBA','BenzigerodeHeimburg_LN',
        'Bell_Beaker_LN','Halberstadt_LBA',"Corded_Ware_LN",'Alberstedt_LN','Samara_HG',
        'Esperstedt_MN','Starcevo_EN','LBKT_EN','Yamnaya','Spain_MN','Spain_EN',
        'Karsdorf_LN','Baalberge_MN']
p1="Corded_Ware_LN"
haak.remove(p1)
f4vals={}
logfile=open("/public/adna/student/2018class/yang_mel/lesson9/CW_f4_%s.log" % (p1),'r')
for line in logfile:
    x=line.split()
    if x[0] != "result:": continue
    if (x[1],x[2]) not in f4vals: f4vals[(x[1],x[2])]=[float(x[5])]
    else: f4vals[(x[1],x[2])].append(float(x[5]))
    #if x[2]=="Karelia_HG": print x
logfile.close()
#for i in f4vals: print i,len(f4vals[i])

for pop1 in haak:
    print pop1
    fig = plt.figure(figsize=(13,10))
    count=1
    #for ind1,pop1 in enumerate(haak):
    for ind2,pop2 in enumerate(haak):
        if pop2==pop1: continue
        ax1=fig.add_subplot(4,4,count)
        count+=1
        #if ind1>=ind2: continue
        ax1.axhline(y=0,ls='dashed', c='black',alpha=0.5)
        ax1.axvline(x=0,ls='dashed', c='black',alpha=0.5)
        ax1.plot(f4vals[(p1,pop1)],f4vals[(p1,pop2)],'o')
        ax1.set_xlabel(pop1)
        ax1.set_ylabel(pop2)

    plt.tight_layout()
    plt.subplots_adjust(wspace=0.5)
    fig.autofmt_xdate()
    #plt.show()
    plt.savefig("/public/adna/student/2018class/yang_mel/lesson9/CW_f4_%s_%s.png" % (p1,pop1))
````
