## Python script to make mapfile with GPS

```python
%%writefile /mnt/solexa/mel_yang/shandong_0118/outf3/makef3mapfile.py

import sys

pd=sys.argv[1] ## /mnt/solexa/mel_yang/shandong_0118/outf3/
fh=sys.argv[2] ## UPA_all9
sgdpanc=sys.argv[3] ## SGDP or ANC
out=sys.argv[4]  ## Mbuti, Altai, Denisovan, Yoruba, Ju_hoan_North, Mota
source1=sys.argv[5] ## Linzi, Han, etc
target=sys.argv[6] ## OUT, OUTEUR
tv=sys.argv[7] ## Y, N

addedsubset="X_%s_S1_%s_S2_%s" % (target,source1,sgdpanc)

GPSfile=open("/mnt/solexa/mel_yang/data/GPS/%s.GPS" % sgdpanc,'r')
mygps={ line.split()[0]:[line.split()[1],line.split()[2]] for line in GPSfile }
GPSfile.close()

myfile=open(pd+fh+".f3."+addedsubset+".log",'r')

for line in myfile:
    x=line.split()
    if x[0]=="result:": pass #x=x[:5]
    else: continue
    if x[3] != out: continue
    ##newfile.write('\t'.join(x)+'\n') ##for log file
    if x[2] in mygps: mygps[x[2]].append(x[4]); mygps[x[2]].append(x[5])
    else: continue
myfile.close()

if tv=="Y": newfile=open(pd+"OUTF3.transv.%s.%s.%s.map" % (out,source1,sgdpanc),'w')
else: newfile=open(pd+"OUTF3.%s.%s.%s.map" % (out,source1,sgdpanc),'w')
for i in mygps: 
    if len(mygps[i]) == 4: newfile.write(i+' '+' '.join(mygps[i])+'\n')
newfile.close()
```

```python
%%writefile /mnt/solexa/mel_yang/shandong_0118/outf3/heatmap_Pontus_meledit_onlyanc.R
args=commandArgs(TRUE)
#PREFIX = as.character(args[1])
PROJDIR = as.character(args[1])
STATFILE = as.character(args[2])
LABEL = as.character(args[3])
mainlat=as.numeric(args[4])
mainlon=as.numeric(args[5])
tv=as.character(args[6])

library(fields)
library(RColorBrewer)

padding = 6

data=read.table(paste(PROJDIR,'OUTF3',tv,'.',STATFILE,'.map',sep=''))

NAMES <- as.character(data[,1])
Latitude<-as.numeric(data[,2])
Longitude<-as.numeric(data[,3])

minLong <- min(Longitude)-5*padding
maxLong <- max(Longitude)+padding
minLat <- min(Latitude)-padding
maxLat <- max(Latitude)+padding

#minLong <- 80-5*padding
#maxLong <- 180.5+padding
#minLat <- -75-padding
#maxLat <- 75+padding

if (grepl('SGDP',STATFILE)) {
    legendposx= minLong+padding
    minLong = -187.8
    maxLong = 180.5
    minLat = -47.3
    maxLat = 75.0
    pointshape=21
    }
if (grepl('ANC',STATFILE)) {
    legendposx= maxLong-2*padding
    minLong = -10
    maxLong = 180.5
    minLat = 10
    maxLat = 75.0
    pointshape=22
    }

statistic=as.numeric(data[,4])
rgb.palette<-colorRampPalette(c("black","midnightblue","blue","lightskyblue","yellow","gold","red")) #
colours<-rgb.palette(30)

x<-statistic
maxstat =max(statistic)
minstat = min(statistic)

png(paste(PROJDIR,'OUTF3.',STATFILE,'.png',sep=''),width=1440*300/72,height=(maxLat-minLat)*1440*300/(maxLong-minLong)/72,res=300)
cols <- as.character(cut(statistic, breaks = length(colours),labels =colours))
plot(Longitude,Latitude,type="n",main=expression(paste(f[3])),xlim=c(minLong,maxLong),ylim=c(minLat,maxLat), cex.lab=2,cex.axis=2)

library(mapdata)
library(maps)
library(maptools)


map(database='worldHires', fill=TRUE,xlim=c(minLong,maxLong),ylim=c(minLat,maxLat),col='lightgray',boundary=FALSE)

CASPIAN <- read.table('/home/mel_yang/rcode/Caspian.txt')
polygon(CASPIAN[,2:1],col='white',border='white')

abline(h = mainlat, lty = 2)
abline(v = mainlon, lty = 2)

PCH <- vector(length=length(Longitude))
PCH[1:length(PCH)] <- pointshape

points(Longitude,Latitude,pch=PCH,bg=cols,col="black",cex=3)
box()


###legend
tempcolours=rev(colours)
#legendposx= 150
#legendposx= maxLong-3*padding
legendwidth=2*padding
#legendtop=40
legendtop = maxLat-(maxLat-minLat)/2
legendbottom= minLat+(maxLat-minLat)/10
legendspan=legendtop-legendbottom
legendincrement=legendspan/length(colours)
for(i in 1:length(tempcolours)) {
  rect(legendposx,legendtop-(legendincrement*i),(legendposx+legendwidth),legendtop-(legendincrement*i-legendincrement),col=tempcolours[i],border=NA)
  #text((legendposx+15),(legendtop-(legendincrement*i-legendincrement/2)),"hello")
}
text(legendposx+legendwidth/2,legendtop+(maxLat-minLat)/20,round(maxstat,3),cex=2)
text(legendposx+legendwidth/2,legendbottom-(maxLat-minLat)/20,round(minstat,3),cex=2)
rect(legendposx,legendtop,(legendposx+legendwidth),legendbottom,col=NA)

if (grepl('F4',STATFILE)) {
text(minLong+(maxLong-minLong)/2.5,minLat+2*padding,labels=expression(f[4]),cex=2,adj=0)
text(minLong+(maxLong-minLong)/2.5+1.25,minLat+2*padding,labels=LABEL,cex=2,adj=0)
}

if (grepl('F3',STATFILE)) {
    text(minLong,maxLat-2,labels=expression(f[3]),cex=1.3,adj=0)
    text(minLong+3,maxLat-2,labels=LABEL,cex=1.3,adj=0)
}
 
dev.off()
```

## Make xlsx file

```python
%%writefile /mnt/solexa/mel_yang/shandong_0118/outf3/makexlsx.py
## python makexlsx.py /mnt/solexa/mel_yang/shandong_0118/outf3/ OUTF3.Mbuti.Dai.SGDP.map
## python formatD.py ${pd} ${fh}

import sys
import xlsxwriter as xls

pd=sys.argv[1]
fh=sys.argv[2]

newfile=xls.Workbook(pd+fh[:-4]+".xlsx")
worksheet=newfile.add_worksheet()
##FORMATS
mynum = newfile.add_format({'num_format': '0.0000','center_across':True})
mystr = newfile.add_format({'align':'center','valign':'vcenter'})
worksheet.set_column(0,20,8)

myfile=open(pd+fh,'r')

for indrow,line in enumerate(myfile):
    x=line.split()
    for indcol,xi in enumerate(x):
        try:
            complex(xi)
            worksheet.write_number(indrow,indcol,float(xi),mynum)
        except ValueError:
            worksheet.write_string(indrow,indcol,xi,mystr)
newfile.close()
myfile.close()
```
