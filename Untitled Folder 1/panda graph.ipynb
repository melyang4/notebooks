```python
##%%writefile /mnt/solexa/mel_yang/panda/FABgraph.py
import numpy as np
import matplotlib.pyplot as plt
import sys

def mkfigure(filename,titlename,xest=False):
    ##Read in data
    myfile=open(filename,'r')
    #myfile=open("/mnt/solexa/mel_yang/panda/data_for_FAB_plot.txt")
    myary=np.zeros((0,3))
    for ind,line in enumerate(myfile):
        if line[0]=="#": mydat=[float(i) for i in line.strip().split(':')[1].split(',')]; continue
        if line[0]=='X': continue
        x=np.array([float(i) for i in line.split()])
        myary=np.vstack((myary,x))
    myfile.close()

    ##Making figure
    fig = plt.figure(figsize=(10,6))
    ax1=fig.add_subplot(1,1,1)

    ## Labels
    ax1.set_ylabel("F(A|B)",fontsize=16)
    ax1.set_xlabel("Split Time (4Ne x gen units)",fontsize=16)
    ax1.set_title(titlename,fontsize=24)

    ## Plot Simulated
    ax1.errorbar(myary[:,0],myary[:,1],yerr=2*myary[:,2],ecolor='black',capsize=3,capthick=1)
    ax1.plot(myary[:,0],myary[:,1],'.-',label="simulated")
    ## Plot Empirical
    emp=mydat[0]
    se=mydat[1]
    emplower,empupper=(emp-se,emp+se)
    ax1.axhline(y=emp,color='green',label="empirical")
    ax1.axhline(y=emplower,linestyle='--',color='green',label="empirical\n+/- 1 S.E.")
    ax1.axhline(y=empupper,linestyle='--',color='green')

    plt.legend(frameon=False)
    ##plt.savefig(sys.argv[1][:-4]+".png")
    if xest==True:
        ##Calculate x range from line based on two closest points to empirical mean
        diff=list(abs(myary[:,1]-emp))
        diffmins=tuple(sorted(diff)[:2])
        minind1,minind2=(diff.index(diffmins[0]),diff.index(diffmins[1]))
        x1,x2=(myary[minind1,0],myary[minind2,0])
        y1,y2=(myary[minind1,1],myary[minind2,1])
        m=(y1-y2)/(x1-x2)
        b=y1-m*x1
        myfit = np.polyval([m,b],myary[:,0])
        ax1.plot(myary[:,0],myfit,'--',color="orange")
        xemp=(emp-b)/m
        xempu=(empupper-b)/m
        xempl=(emplower-b)/m
        print "Empirical_x:", xemp
        print "Lower_x:", xempu
        print "Upper_x:", xempl
        ax1.axvline(x=xemp,linestyle='-',color='orange')
        ax1.axvline(x=xempu,linestyle='--',color='orange')
        ax1.axvline(x=xempl,linestyle='--',color='orange')
        plt.savefig(sys.argv[1][:-4]+".xest.png")
    else: 
        plt.savefig(sys.argv[1][:-4]+".png")
        #plt.show()
        
    return "Completed making figure"

mkfigure(sys.argv[1],"F(LafrC | LafrB)",xest=False)
mkfigure(sys.argv[1],"F(LafrC | LafrB)",xest=True)
```
