```python
from plateaumodules import piid1, parseoutf3
import collections
import xlsxwriter as xls
agq=["Zongri1a","Zongri23","Shangluzhuang","Hejiatai","Heodong","Daxiang","Xiahaishi",
     "Hupo","Sanheyi","Qunke","Lajigai","Wenpuju","Jinchankou","Qijiaping",
     "Wuba","Huoshaogou","Lierbao"]
anx=["Pukagongma","Pukagongma7","Yushu","Redilong","Xiaoenda","Butaxiongqu","Caixiutang",
    "Chokhopani","Mebrak","Samdzong"]

popname,tv,tv1,myinds=('aGQ','.tv.',"Transversions only",agq)
#popname,tv,tv1,myinds=('aNX','.tv.',"Transversions only",anx)
#popname,tv,tv1,myinds=('aGQ','.',"All SNPs",agq)
#popname,tv,tv1,myinds=('aNX','.',"All SNPs",anx)

pd="/mnt/solexa/mel_yang/aPlateau/outf3/"
myf3dat1=parseoutf3(pd+"%s-panel-mbuti%sf3.log" % ("aGQ",tv))
myf3dat2=parseoutf3(pd+"%s-panel-mbuti%sf3.log" % ("aNX",tv))
myf3dat1.update(myf3dat2)
out="Mbuti"
aggdat=collections.defaultdict(list)
for ind1,agqind in enumerate(agq): 
    for ind2,x in enumerate(agq):
        if ind1>=ind2: continue
        try:
            aggdat[(agqind,x,out)]=myf3dat1[agqind][(agqind,x,out)]
        except KeyError:
            aggdat[(agqind,x,out)]=myf3dat1[x][(x,agqind,out)]
    for ind2,x in enumerate(anx):
        if ind1>=ind2: continue
        try:
            aggdat[(agqind,x,out)]=myf3dat1[agqind][(agqind,x,out)]
        except KeyError:
            aggdat[(agqind,x,out)]=myf3dat1[x][(x,agqind,out)]
for ind1,agqind in enumerate(anx): 
    for ind2,x in enumerate(anx):
        if ind1>=ind2: continue
        try:
            aggdat[(agqind,x,out)]=myf3dat1[agqind][(agqind,x,out)]
        except KeyError:
            aggdat[(agqind,x,out)]=myf3dat1[x][(x,agqind,out)]
            
            
newfile=xls.Workbook(pd+"agq_anx_outf3.xlsx")
worksheet=newfile.add_worksheet()
agqall = newfile.add_format({'center_across':True,'bg_color':'#B0E0E6'})
anxall = newfile.add_format({'center_across':True,'bg_color':'yellow'})
agqanx = newfile.add_format({'center_across':True,'bg_color':'#98FB98'})
worksheet.set_column(1,7,12)

row=0
for mykey in aggdat:
    row+=1
    if mykey[0] in agq and mykey[1] in agq:
        worksheet.write_string(row,1,mykey[0],agqall)
        worksheet.write_string(row,2,mykey[1],agqall)
        worksheet.write_string(row,3,mykey[2],agqall)
        for i in range(4): worksheet.write_number(row,i+4,aggdat[mykey][i],agqall)
    elif mykey[0] in agq and mykey[1] in anx:
        worksheet.write_string(row,1,mykey[0],agqanx)
        worksheet.write_string(row,2,mykey[1],agqanx)
        worksheet.write_string(row,3,mykey[2],agqanx)
        for i in range(4): worksheet.write_number(row,i+4,aggdat[mykey][i],agqanx)
    elif mykey[1] in agq and mykey[0] in anx:
        worksheet.write_string(row,1,mykey[0],agqanx)
        worksheet.write_string(row,2,mykey[1],agqanx)
        worksheet.write_string(row,3,mykey[2],agqanx)
        for i in range(4): worksheet.write_number(row,i+4,aggdat[mykey][i],agqanx)
    elif mykey[0] in anx and mykey[1] in anx:
        worksheet.write_string(row,1,mykey[0],anxall)
        worksheet.write_string(row,2,mykey[1],anxall)
        worksheet.write_string(row,3,mykey[2],anxall)
        for i in range(4): worksheet.write_number(row,i+4,aggdat[mykey][i],anxall)

newfile.close()
```
