```python
import numpy as np
import plateaumodules as pm
import matplotlib.pyplot as plt
import matplotlib
pD="/mnt/solexa/mel_yang/aPlateau/"
fh="aPlateau.dstat.log"
tv='all'

agq=["Zongri1a","Zongri23","Shangluzhuang","Hejiatai","Heodong","Daxiang","Xiahaishi",
     "Hupo","Sanheyi","Qunke","Lajigai","Wenpuju","Jinchankou","Qijiaping",
     "Wuba","Huoshaogou","Lierbao","Shannashuzha",][::-1]
wantedcolors=["blue"]*len(agq)
myxmin=[-0.08]*2
myxmax=[0.05]*2
mysubplot=0
fig = plt.figure(figsize=(10,5))

popA=agq
popB=["Jinchankou"]
varypop=["Tianyuan",'Vestonice16']
myind="Han"
outs=["Mbuti","Chimp"]
for count,out in enumerate(outs):
    #myind="Shangluzhuang"
    #myind="Hejiatai"
    
    mypops1=[popA,myind,popB,out]
    mypops2=[popA,popB,myind,out]
    DorZ="Z"

    Zary = pm.mkDary(pD,fh,mypops2,(2,3),(0,1),'Z')
    Dary = pm.mkDary(pD,fh,mypops2,(2,3),(0,1),'D')
    mylabel="D(%s,%s;%s,%s)" % ('aGQ',popB[0],myind,out)

    SEary=Dary/Zary 
    mynames=agq
    
    ax1=fig.add_subplot(1,len(varypop),count+1)
    for ind,i in enumerate(agq):
        #print tyD[i]
        #print Dary[ind][0]
        ax1.errorbar(Dary[ind][0],ind,xerr=SEary[ind][0],color=wantedcolors[ind],markeredgecolor="black",markersize=10,ecolor='darkgrey',fmt='o',capthick=0,linewidth=7)
        ax1.errorbar(Dary[ind][0],ind,xerr=3*SEary[ind][0],color=wantedcolors[ind],markeredgecolor="black",markersize=10,ecolor='darkgrey',fmt='o',capthick=0,linewidth=4)
        ax1.plot(Dary[ind],ind,'.')
    #ax1.set_yticks(len(mynames))
    myys = np.array(range(len(mynames)))
    ax1.set_yticks(myys)
    if count==0:
        ax1.set_yticklabels(mynames,fontsize=16)
    else: 
        ax1.set_yticklabels([' ']*len(mynames),fontsize=3)

    ax1.set_ylim(-1,len(mynames))
    ax1.set_xlim(myxmin[count],myxmax[count])

    xlabel = ax1.get_xticks().tolist()
    ax1.set_xticklabels(xlabel,rotation=45,fontsize=15)

    ax1.set_title(mylabel,fontsize=15)

    ax1.axvline(x=0, ls='dashed', c='black')
    for myline in [3.5]:
        ax1.axhline(y=myline, ls='solid', c='black',alpha=0.5)

    ax1.set_xlabel("D, %s" % tv,fontsize=15)
    
plt.tight_layout()
plt.show()
```

## Heatmap

```python
import numpy as np
import plateaumodules as pm
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.colors
pD="/mnt/solexa/mel_yang/aPlateau/"
fh="aPlateau.dstat.log"
tv='all'
fs=15
agq=["Zongri1a","Zongri23","Shangluzhuang","Hejiatai","Heodong","Daxiang","Xiahaishi",
     "Hupo","Sanheyi","Qunke","Lajigai","Wenpuju","Jinchankou","Qijiaping",
     "Wuba","Huoshaogou","Lierbao","Shannashuzha",]
wantedcolors=["blue"]*len(agq)
header1,header2=agq,agq
popA=agq
popB=agq
varypop=["Tianyuan",'Vestonice16']
myind="Sherpa"
outs=["Mbuti","Chimp"]
for count,out in enumerate(outs):
    #myind="Shangluzhuang"
    #myind="Hejiatai"
    
    mypops1=[popA,myind,popB,out]
    mypops2=[popA,popB,myind,out]
    DorZ="Z"

    Zary = pm.mkDary(pD,fh,mypops2,(2,3),(0,1),'Z')
    Dary = pm.mkDary(pD,fh,mypops2,(2,3),(0,1),'D')
    mylabel="D(%s,%s;%s,%s)" % ('P1',"P2",myind,out)
    Zary = np.nan_to_num(Zary)
    
    levels = range(-10,11)
    multiplier=((len(levels)-6)/2)+1
    colors = ['#333399']*multiplier + ['#CCCCFF'] + ['white']*4 + ['#FFEFD5'] + ['orange']*multiplier
    cmap1,norm = matplotlib.colors.from_levels_and_colors(levels, colors,extend='both')
    
    fig,ax = plt.subplots(figsize=(10,10))
    heatmap = ax.pcolor(Zary ,cmap=cmap1, vmin=-10,vmax=10)
    cbar=plt.colorbar(heatmap)
    
    ##LEGEND
    cbar.ax.text(.5, 0.25, '$D<0$', color='black', ha='center', va='center',fontsize=fs-22)
    cbar.ax.text(.5, 0.5, '$N.S.$', color='black', ha='center', va='center',fontsize=fs-22)
    cbar.ax.text(.5, 0.75, '$D>0$', color='black', ha='center', va='center',fontsize=fs-22)
    cbar.ax.get_yaxis().labelpad = 15
    cbar.ax.set_ylabel('Z-score', rotation=270,fontsize=fs)
    
    ax.grid(True)
    ax.set_xticks(np.arange(len(header1)), minor=False)
    ax.set_yticks(np.arange(len(header2)), minor=False)
    
    column_labels = header1
    row_labels = header2
    ax.set_xlim((0,len(header1)))
    ax.set_ylim((0,len(header2)))
    ax.set_xticklabels(column_labels, minor=False,fontsize=fs,rotation=75,ha='left')
    ax.set_yticklabels(row_labels, minor=False,fontsize=fs,va='top')
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    ax.text(-3.5,-0.5,"P1/P2",fontsize=fs+5)
    
plt.show()
print myind
```
